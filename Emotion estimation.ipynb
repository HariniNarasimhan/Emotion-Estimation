```python
import cv2
from mtcnn.mtcnn import MTCNN
import face_alignment
from skimage import io

fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D,flip_input=False)

# input = np.reshape(fer['pixels'].iloc[3].split(' '), (48,48)).astype('float')
```

```python
import matplotlib.pyplot as plt
import pickle
from sklearn.externals import joblib
import numpy as np
import pandas as pd
import os
from keras.models import load_model
from keras.preprocessing.image import img_to_array
```

```python
emotion_classifier = load_model('<emotion_estimation_model.h5>', compile=False)
EMOTIONS = ["angry" ,"disgust","scared", "happy", "sad", "surprised",
 "neutral"]
```

```python
detector = MTCNN()
```

```python
cap = cv2.VideoCapture('<input video file>')
while True: 
#     print('reading')
    #Capture frame-by-frame
    __, frame = cap.read()
    #Use MTCNN to detect faces
    result = detector.detect_faces(frame)
    if result != []:
        for person in result:
            bounding_box = person['box']
            keypoints = person['keypoints']
    
            cv2.rectangle(frame,
                          (bounding_box[0], bounding_box[1]),
                          (bounding_box[0]+bounding_box[2], bounding_box[1] + bounding_box[3]),
                          (0,155,255),
                          2)
            
            img = cv2.cvtColor(frame[bounding_box[1]:bounding_box[1] + bounding_box[3],
                                     bounding_box[0]:bounding_box[0]+bounding_box[2]],cv2.COLOR_BGR2GRAY)
            if img is not None:
                roi = cv2.resize(img, (64, 64))
                roi = roi.astype("float") / 255.0
                roi = img_to_array(roi)
                roi = np.expand_dims(roi, axis=0)


                preds = emotion_classifier.predict(roi)[0]
                emotion_probability = np.max(preds)
                label = EMOTIONS[preds.argmax()]
    #             cv2.circle(frame,(keypoints['left_eye']), 2, (0,155,255), 2)
    #             cv2.circle(frame,(keypoints['right_eye']), 2, (0,155,255), 2)
    #             cv2.circle(frame,(keypoints['nose']), 2, (0,155,255), 2)
    #             cv2.circle(frame,(keypoints['mouth_left']), 2, (0,155,255), 2)
    #             cv2.circle(frame,(keypoints['mouth_right']), 2, (0,155,255), 2)
    #             result = linear_svc.predict(points)
                cv2.putText(frame, label, (bounding_box[0],bounding_box[1]), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 255), 2)
            
    #display resulting frame
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) &0xFF == ord('q'):
        break
#When everything's done, release capture
cap.release()
# out.release()
cv2.destroyAllWindows()
```
